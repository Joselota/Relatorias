# 🔁 Control de Flujo en Python
Este notebook cubre las estructuras lógicas que permiten **tomar decisiones** y **repetir acciones** en Python.
---

## 1. Condicionales (`if`, `elif`, `else`)
Ejecutan un bloque de código si se cumple una condición.

```python
edad = 18
if edad >= 18:
    print("Es mayor de edad")
elif edad >= 13:
    print("Es adolescente")
else:
    print("Es menor de edad")
```

## 2. Ciclo `while`
Repite instrucciones mientras la condición sea verdadera.

```python
x = 0
while x < 3:
    print("Repetición", x)
    x += 1
```

## 3. Ciclo `for`
Recorre los elementos de una lista o secuencia.

```python
frutas = ["manzana", "pera", "uva"]
for fruta in frutas:
    print(fruta)
```

## 4. Uso de `break` y `continue`
- `break` termina el ciclo.
- `continue` salta a la siguiente iteración.

```python
print("Ejemplo con break")
for i in range(5):
    if i == 3:
        break
    print(i)

print("\nEjemplo con continue")
for i in range(5):
    if i == 2:
        continue
    print(i)
```
