## √Årbol de probabilidades1: Situaci√≥n: Lanzas una moneda dos veces.

```python
from graphviz import Digraph

dot = Digraph("Moneda", format="png")

dot.attr(rankdir="TB", fontname="Arial", bgcolor="white")
dot.attr('node', fontname="Arial", style="filled,rounded")
dot.attr('edge', fontname="Arial", fontsize="10")

dot.node("I", "INICIO", fillcolor="#3498DB", fontcolor="white", shape="circle")

dot.node("C1", "Cara", fillcolor="#27AE60", fontcolor="white")
dot.node("S1", "Sello", fillcolor="#E74C3C", fontcolor="white")

dot.node("CC", "Cara-Cara\nP=25%", fillcolor="#1ABC9C", fontcolor="white")
dot.node("CS", "Cara-Sello\nP=25%", fillcolor="#16A085", fontcolor="white")
dot.node("SC", "Sello-Cara\nP=25%", fillcolor="#C0392B", fontcolor="white")
dot.node("SS", "Sello-Sello\nP=25%", fillcolor="#A93226", fontcolor="white")

dot.edge("I", "C1", label="50%", color="#27AE60")
dot.edge("I", "S1", label="50%", color="#E74C3C")

dot.edge("C1", "CC", label="50%", color="#1ABC9C")
dot.edge("C1", "CS", label="50%", color="#16A085")
dot.edge("S1", "SC", label="50%", color="#C0392B")
dot.edge("S1", "SS", label="50%", color="#A93226")

dot
```

## Ejemplo 2: √Årbol de decisiones simple llevar paraguas

### Ejemplo: Flujo de proceso

## EJEMPLO: Pron√≥stico del Tiempo

```python
from graphviz import Digraph

# -----------------------------
# EJEMPLO: Pron√≥stico del Tiempo ‚òî
# -----------------------------

# Probabilidades del clima
p_lluvia = 0.70
p_no_lluvia = 0.30

# Si LLUEVE
p_paraguas_si_llueve = 0.80
p_no_paraguas_si_llueve = 0.20

# Si NO LLUEVE
p_paraguas_si_no_llueve = 0.10
p_no_paraguas_si_no_llueve = 0.90

# -----------------------------
# Probabilidades conjuntas (hojas del √°rbol)
# -----------------------------
p_lluvia_y_paraguas = p_lluvia * p_paraguas_si_llueve
p_lluvia_y_no_paraguas = p_lluvia * p_no_paraguas_si_llueve
p_no_lluvia_y_paraguas = p_no_lluvia * p_paraguas_si_no_llueve
p_no_lluvia_y_no_paraguas = p_no_lluvia * p_no_paraguas_si_no_llueve

# -----------------------------
# Construcci√≥n del √°rbol
# -----------------------------
dot = Digraph("ArbolClima", format="png")

# Configuraci√≥n general
dot.attr(rankdir="TB",
         fontname="Arial",
         fontsize="14",
         splines="ortho",
         bgcolor="#F0F8FF",
         nodesep="1.2",
         ranksep="1.2")

dot.attr('node',
         fontname="Arial Bold",
         fontsize="12",
         style="filled,rounded",
         margin="0.4,0.3")

dot.attr('edge',
         fontname="Arial",
         fontsize="12",
         fontcolor="#2C3E50",
         penwidth="3")

# Nodo inicial
dot.node("Inicio", "üå§Ô∏è MA√ëANA",
         shape="circle",
         fillcolor="#3498DB",
         fontcolor="white",
         width="1.5",
         fontsize="16")

# Primer nivel: CLIMA
dot.node("Llueve", "‚òî LLUEVE\n\nP(Lluvia) = 70%",
         shape="box",
         fillcolor="#5DADE2",
         fontcolor="white",
         style="filled,rounded",
         width="2")

dot.node("NoLlueve", "‚òÄÔ∏è NO LLUEVE\n\nP(No lluvia) = 30%",
         shape="box",
         fillcolor="#F39C12",
         fontcolor="white",
         style="filled,rounded",
         width="2")

# Segundo nivel: DECISI√ìN DE PARAGUAS (si llueve)
dot.node("LP", "üåÇ USO PARAGUAS\n\nP(‚òÇ|‚òî) = 80%\n\nP(‚òî‚à©‚òÇ) = 56%",
         shape="box",
         fillcolor="#1ABC9C",
         fontcolor="white",
         style="filled,rounded")

dot.node("LNP", "üö∂ NO USO PARAGUAS\n\nP(no ‚òÇ|‚òî) = 20%\n\nP(‚òî‚à©no ‚òÇ) = 14%",
         shape="box",
         fillcolor="#E74C3C",
         fontcolor="white",
         style="filled,rounded")

# Segundo nivel: DECISI√ìN DE PARAGUAS (si NO llueve)
dot.node("NLP", "üåÇ USO PARAGUAS\n\nP(‚òÇ|‚òÄ) = 10%\n\nP(‚òÄ‚à©‚òÇ) = 3%",
         shape="box",
         fillcolor="#95A5A6",
         fontcolor="white",
         style="filled,rounded")

dot.node("NLNP", "üö∂ NO USO PARAGUAS\n\nP(no ‚òÇ|‚òÄ) = 90%\n\nP(‚òÄ‚à©no ‚òÇ) = 27%",
         shape="box",
         fillcolor="#F1C40F",
         fontcolor="#2C3E50",
         style="filled,rounded")

# Aristas
dot.edge("Inicio", "Llueve", label="  70%  ", color="#5DADE2", penwidth="4")
dot.edge("Inicio", "NoLlueve", label="  30%  ", color="#F39C12", penwidth="4")

dot.edge("Llueve", "LP", label="  80%  ", color="#1ABC9C", penwidth="3")
dot.edge("Llueve", "LNP", label="  20%  ", color="#E74C3C", penwidth="3")

dot.edge("NoLlueve", "NLP", label="  10%  ", color="#95A5A6", penwidth="3")
dot.edge("NoLlueve", "NLNP", label="  90%  ", color="#F1C40F", penwidth="3")

dot
```
