| Columna                     | ¿Qué representa?                 | Primeras ideas de uso                               |
| --------------------------- | -------------------------------- | --------------------------------------------------- |
| `DIM_COUNTRY_CODE`          | Código ISO-3 del país            | filtrar un país, agrupar por región…                |
| `DIM_YEAR_CODE`             | Año (numérico)                   | series de tiempo                                    |
| `DIM_GHECAUSE_TITLE`        | Causa de muerte (clasif. GHE)    | ranking de causas, filtros temáticos                |
| `DIM_SEX_CODE`              | Sexo (`BTSX`, etc.)              | comparar por sexo o usar “ambos sexos”              |
| `VAL_DTHS_RATE100K_NUMERIC` | Tasa de muertes por 100 000 hab. | métrica principal, apta para promedios, mapas, etc. |


```python
# Países disponibles (primeros 10)
df['DIM_COUNTRY_CODE'].unique()[:10]
```

```python
# Años cubiertos
df['DIM_YEAR_CODE'].min(), df['DIM_YEAR_CODE'].max()
```

```python
# Top 5 causas más frecuentes en el dataset
(
    df['DIM_GHECAUSE_TITLE']
      .value_counts()
      .head(5)
)
```

```python
#Seleccionemos solo los registros de CHILE (CHL)
# Filtrar por código ISO-3 de Chile
df_cl = df[df["DIM_COUNTRY_CODE"] == "CHL"].copy()

# Comprobar tamaño resultante
df_cl.shape
```

```python
# ¿De qué rango de años disponemos?
# 1) Años cubiertos en df_cl
df_cl["DIM_YEAR_CODE"].unique()
```

```python
# ¿Queremos quedarnos con “ambos sexos” (BTSX) o comparar hombres/mujeres?
# 2) Sexos disponibles
df_cl["DIM_SEX_CODE"].unique()
```

```python
# ¿Alguna causa de muerte te interesa destacar primero (por ejemplo, las 3 más altas)?
# 3) Causas más frecuentes (solo ver 5)
df_cl["DIM_GHECAUSE_TITLE"].value_counts().head()
```

```python
# cuáles son las 10 causas con mayor tasa de muertes por 100 000 habitantes
# ordenar de mayor a menor y tomar las 10 primeras
top10_causas = (
    df_cl.sort_values("VAL_DTHS_RATE100K_NUMERIC", ascending=False)
         .head(10)
)

top10_causas
```

```python
import matplotlib.pyplot as plt

# Ordenamos de mayor a menor
ordenado_cl = (
    df_cl.sort_values("VAL_DTHS_RATE100K_NUMERIC", ascending=False)
         .reset_index(drop=True)
)

fig, ax = plt.subplots(figsize=(14, 24))      # ↑ alto más grande para 134 barras
barras = ax.barh(
    ordenado_cl["DIM_GHECAUSE_TITLE"],
    ordenado_cl["VAL_DTHS_RATE100K_NUMERIC"]
)

# Etiquetamos cada barra con la tasa (sin decimales extra)
ax.bar_label(barras, fmt='%.2f', padding=3)

ax.set_title("Chile · Tasas de mortalidad por causa (2021, por 100 000 hab.)")
ax.set_xlabel("Tasa de muertes")
ax.set_ylabel("Causa GHE")
ax.tick_params(axis='y', labelsize=8)
ax.invert_yaxis()            # mayor arriba
plt.tight_layout()
plt.show()
```

```python
# Los 10 primeros
import matplotlib.pyplot as plt

# 1️⃣ Filtramos y ordenamos (ya lo tenías, pero lo dejo completo)
top10_cl = (
    df_cl.sort_values("VAL_DTHS_RATE100K_NUMERIC", ascending=False)
          .head(10)
)

# 2️⃣ Gráfico de barras horizontales
fig, ax = plt.subplots(figsize=(10, 6))
barras = ax.barh(
    top10_cl["DIM_GHECAUSE_TITLE"],
    top10_cl["VAL_DTHS_RATE100K_NUMERIC"]
)

# 3️⃣ Etiquetas con la cifra (un decimal)
ax.bar_label(barras, fmt='%.1f', padding=3)

ax.set_title("Chile · Top 10 causas de mortalidad (2021)")
ax.set_xlabel("Tasa de muertes por 100 000 hab.")
ax.set_ylabel("Causa (GHE)")
ax.invert_yaxis()           # la mayor arriba
plt.tight_layout()
plt.show()
```

# Grafico a nivel global

```python
df_21 = df[
    (df["DIM_YEAR_CODE"] == 2021) &
    (df["DIM_SEX_CODE"] == "BTSX")
].copy()

df_21.shape          # ¿cuántas filas te muestra?
```

```python
# ranking mundial tomando la media de la tasa de cada causa
#Ranking
ranking_mundial = (
    df.groupby("DIM_GHECAUSE_TITLE")["VAL_DTHS_RATE100K_NUMERIC"]
      .mean()                         # ⬅️  aquí usamos “promedio”
      .sort_values(ascending=False)
      .head(10)                       # top-10
      .reset_index()
)

ranking_mundial
```

```python
import matplotlib.pyplot as plt

# ranking_mundial ya tiene las columnas:
#   DIM_GHECAUSE_TITLE  – nombre de la causa
#   VAL_DTHS_RATE100K_NUMERIC  – tasa media global

fig, ax = plt.subplots(figsize=(10, 6))
barras = ax.barh(
    ranking_mundial["DIM_GHECAUSE_TITLE"],
    ranking_mundial["VAL_DTHS_RATE100K_NUMERIC"]
)

# Etiquetas con un decimal
ax.bar_label(barras, fmt='%.1f', padding=3)

ax.set_title("Top 10 causas de mortalidad – promedio global 2021")
ax.set_xlabel("Tasa media de muertes por 100 000 habitantes")
ax.set_ylabel("Causa (GHE)")
ax.invert_yaxis()        # mayor arriba
plt.tight_layout()
plt.show()
```

```python
#ranking por suma total
ranking_sum = (
    df.groupby("DIM_GHECAUSE_TITLE")["VAL_DTHS_RATE100K_NUMERIC"]
      .sum()                         # ⬅️ ahora usamos sum()
      .sort_values(ascending=False)
      .head(10)                      # top-10
      .reset_index()
)

ranking_sum
```

```python
#ranking por suma total en un gráfico
import matplotlib.pyplot as plt

# ranking_sum ya lo tienes del paso anterior
fig, ax = plt.subplots(figsize=(10, 6))
barras = ax.barh(
    ranking_sum["DIM_GHECAUSE_TITLE"],
    ranking_sum["VAL_DTHS_RATE100K_NUMERIC"]
)

ax.bar_label(barras, fmt='%.0f', padding=3)   # sin decimales, son cifras grandes
ax.set_title("Top 10 causas de mortalidad – suma global 2021")
ax.set_xlabel("Suma de tasas (muertes por 100 000 hab.)")
ax.set_ylabel("Causa (GHE)")
ax.invert_yaxis()
plt.tight_layout()
plt.show()
```
