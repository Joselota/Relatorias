| Columna                  | Ejemplo                      | Qué representa / para qué sirve                                                                                                                       |
| ------------------------ | ---------------------------- | ----------------------------------------------------------------------------------------------------------------------------------------------------- |
| `IND_ID`                 | `90E2E48WHOSIS_000001`       | Identificador interno único de la observación. Suele ser la concatenación del UUID del indicador (`IND_UUID`) y el código del indicador (`IND_CODE`). |
| `IND_CODE`               | `WHOSIS_000001`              | Código oficial del indicador en el repositorio de la OMS (WHO). Sirve como llave cuando quieres unir con metadatos externos del indicador.            |
| `IND_UUID`               | `90E2E48`                    | Identificador corto (tipo UUID abreviado) del indicador. Es estable entre distintas versiones del dataset.                                            |
| `IND_PER_CODE`           | `WHOSIS_000001`              | Código “período-indicador”. En este archivo coincide con `IND_CODE`; en otros conjuntos puede cambiar si el indicador tiene varias periodicidades.    |
| `DIM_TIME`               | `2000`                       | Valor de la dimensión temporal. Aquí es el **año** del dato.                                                                                          |
| `DIM_TIME_TYPE`          | `YEAR`                       | Tipo de unidad temporal (`YEAR`, `MONTH`, etc.). Aquí siempre `YEAR`.                                                                                 |
| `DIM_GEO_CODE_M49`       | `688`                        | Código numérico M49 de la ONU para el territorio (ej. 688 = Serbia). Muy útil para unir con shapefiles o tablas de población.                         |
| `DIM_GEO_CODE_TYPE`      | `COUNTRY`                    | Tipo de entidad geográfica (país, región, etc.). En este archivo todos son `COUNTRY`.                                                                 |
| `DIM_PUBLISH_STATE_CODE` | `PUBLISHED`                  | Estado de publicación del dato (`PUBLISHED`, `UNVERIFIED`, etc.). Ayuda a filtrar versiones preliminares.                                             |
| `IND_NAME`               | `Life expectancy (at birth)` | Nombre legible del indicador. Es lo que mostrarías en un gráfico o título.                                                                            |
| `GEO_NAME_SHORT`         | `Serbia`                     | Nombre corto del territorio al que se refiere la observación.                                                                                         |
| `DIM_SEX`                | `FEMALE`                     | Sexo o grupo poblacional (`FEMALE`, `MALE`, `BTSX` = ambos sexos).                                                                                    |
| `AMOUNT_N`               | `74.334559`                  | **Valor numérico** del indicador (en este caso, esperanza de vida en años). Esta es la columna que normalmente graficarás o resumirás.                |


# Tendencia de Esperanza de Vida en **Chile**

```python
# Filtrar a Chile y al indicador de esperanza de vida al nacer
cl_le = df[
    (df["GEO_NAME_SHORT"] == "Chile") &
    (df["IND_NAME"] == "Life expectancy (at birth)")
].copy()

# ¿Qué años trae?
cl_le["DIM_TIME"].unique()
```

```python
# ¿Qué sexos hay en esas filas de Chile-Life Expectancy?
cl_le["DIM_SEX"].unique()
```

```python
# 1. Seleccionar solo las columnas relevantes
le_cl = cl_le[["DIM_TIME", "DIM_SEX", "AMOUNT_N"]].copy()

# 2. Pivotear: años en filas, sexos en columnas
le_pivot = (
    le_cl.pivot(index="DIM_TIME", columns="DIM_SEX", values="AMOUNT_N")
          .sort_index()          # ordena los años de menor a mayor
)

le_pivot.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

# Dibujamos cada serie
plt.plot(le_pivot.index, le_pivot["FEMALE"], label="Mujeres")
plt.plot(le_pivot.index, le_pivot["MALE"],   label="Hombres")
plt.plot(le_pivot.index, le_pivot["TOTAL"],  label="Total (ambos sexos)", linewidth=2)

# Detalles de formato
plt.title("Chile · Esperanza de vida al nacer (2000-2021)")
plt.xlabel("Año")
plt.ylabel("Años de vida")
plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

# Brecha de género
¿Cuántos años de diferencia hay entre mujeres y hombres a lo largo del tiempo? ¿Ha crecido o disminuido?

```python
#Crear una serie brecha = FEMALE – MALE, describirla (.describe()), graficar línea o barras.
#Primero crearemos la serie con la diferencia año a año:
# Calcula brecha = esperanza de vida mujeres – hombres
brecha = le_pivot["FEMALE"] - le_pivot["MALE"]

# Veamos las 5 primeras filas para confirmar
brecha.head()

```

```python
# Estadística básica de la brecha
brecha.describe()

# Gráfico de la brecha en el tiempo
import matplotlib.pyplot as plt
plt.figure(figsize=(8,4))
plt.plot(brecha.index, brecha, marker="o")
plt.title("Chile · Brecha de esperanza de vida (♀ − ♂)")
plt.xlabel("Año")
plt.ylabel("Años de diferencia")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Año y valor de la menor brecha
brecha.idxmin(), brecha.min()
```

¿Por qué podría haberse acortado?

Mejoras en salud cardiovascular masculina – Chile ha fortalecido programas de pesquisa y tratamiento, lo que reduce mortalidad prematura en hombres.

Reducción paulatina del hábito de fumar – la prevalencia de tabaco ha caído más rápido en varones que en mujeres, cerrando la brecha de riesgos respiratorios.

Menor mortalidad por lesiones externas – siniestros viales y violencia afectan más a hombres; si esos indicadores bajan, la esperanza de vida masculina sube y se acerca.

# Comparar con otro país (Perú)

```python
# 1️⃣ Filtramos a Perú y al indicador de esperanza de vida
pe_le = df[
    (df["GEO_NAME_SHORT"] == "Peru") &
    (df["IND_NAME"] == "Life expectancy (at birth)")
].copy()

# 2️⃣ ¿Qué sexos hay?
pe_le["DIM_SEX"].unique()
```

```python
# Paso 1 · Pivot igual que antes
pe_pivot = (
    pe_le[["DIM_TIME", "DIM_SEX", "AMOUNT_N"]]
      .pivot(index="DIM_TIME", columns="DIM_SEX", values="AMOUNT_N")
      .sort_index()
)

# Paso 2 · Calcular la brecha Perú (♀ – ♂)
brecha_pe = pe_pivot["FEMALE"] - pe_pivot["MALE"]

# Rápido vistazo
brecha_pe.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(9,5))

# Brecha Chile (ya la tenías en `brecha`)
plt.plot(brecha.index, brecha, marker="o", label="Chile")

# Brecha Perú
plt.plot(brecha_pe.index, brecha_pe, marker="o", label="Perú")

plt.title("Brecha de esperanza de vida (♀ − ♂)")
plt.xlabel("Año")
plt.ylabel("Años de diferencia")
plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

# **Volver a utilizar el dataset de enfermedades, consultando en especifico para Perú**

```python
import pandas as pd

# 1. Ruta al archivo ─ cámbiala si no está en la misma carpeta
PATH = "GHE_FULL_DD.csv"

# 2. Leemos el CSV completo  (≈ 24 k filas, 5 columnas)
df = pd.read_csv(PATH)

# 3. --- FILTROS BÁSICOS --------------------------------------------------
#    - País: Perú  (ISO-3 = 'PER')
#    - Año : 2021  (ajusta aquí si quieres otro año)
#    - Sexo: BTSX  (ambos sexos; cambia a 'MALE' o 'FEMALE' si lo necesitas)
mask = (
    (df["DIM_COUNTRY_CODE"] == "PER") &
    (df["DIM_YEAR_CODE"]   == 2021) &
    (df["DIM_SEX_CODE"]    == "BTSX")
)

peru_2021 = df.loc[mask].copy()

# 4. Ordenamos de mayor a menor tasa y nos quedamos con las 10 primeras causas
top10_peru = (
    peru_2021.sort_values("VAL_DTHS_RATE100K_NUMERIC", ascending=False)
             .head(10)
           [["DIM_GHECAUSE_TITLE", "VAL_DTHS_RATE100K_NUMERIC"]]
           .reset_index(drop=True)
)

# 5. Vistazo rápido
print("Top-10 causas de mortalidad Perú 2021 (tasa por 100 000 hab.):")
print(top10_peru.to_string(index=False))
```

```python
import matplotlib.pyplot as plt
plt.barh(top10_peru["DIM_GHECAUSE_TITLE"], top10_peru["VAL_DTHS_RATE100K_NUMERIC"])
plt.gca().invert_yaxis(); plt.tight_layout(); plt.show()
```

# **Comparemos con CHILE y PERU**

```python
import pandas as pd
import matplotlib.pyplot as plt

# === 1. Cargar el CSV completo ===
PATH = "GHE_FULL_DD.csv"          # ajusta si tu ruta es distinta
df = pd.read_csv(PATH)

# === 2. Definir los parámetros de filtrado ===
PAISES  = ["PER", "CHL"]          # ISO-3 de Perú y Chile
ANIO    = 2021
SEXO    = "BTSX"                  # ambos sexos

mask = (
    df["DIM_COUNTRY_CODE"].isin(PAISES) &
    (df["DIM_YEAR_CODE"] == ANIO)      &
    (df["DIM_SEX_CODE"]  == SEXO)
)

df_pc = df.loc[mask, ["DIM_COUNTRY_CODE",
                      "DIM_GHECAUSE_TITLE",
                      "VAL_DTHS_RATE100K_NUMERIC"]].copy()

# === 3. Pivotear a formato ancho (filas = causa, columnas = país) ===
pivot_pc = (
    df_pc.pivot_table(index="DIM_GHECAUSE_TITLE",
                      columns="DIM_COUNTRY_CODE",
                      values="VAL_DTHS_RATE100K_NUMERIC",
                      aggfunc="first")          # solo hay un valor por causa
    .dropna()                                  # descarta causas faltantes en alguno
)

# === 4. (Opcional) Ordenar por la suma de ambos países y tomar Top-10 ===
top10 = (
    pivot_pc.assign(TOTAL=lambda d: d.sum(axis=1))
            .sort_values("TOTAL", ascending=False)
            .drop(columns="TOTAL")
            .head(10)
)

print("\nTop-10 causas (tasa por 100 000 hab.) – 2021")
print(top10)

# === 5. Gráfico comparativo de barras horizontales ===
fig, ax = plt.subplots(figsize=(10, 6))
top10.plot(kind="barh", ax=ax)      # columnas = países, barras agrupadas
ax.set_title("Perú vs. Chile · Top 10 causas de mortalidad (2021, ambos sexos)")
ax.set_xlabel("Tasa de muertes por 100 000 habitantes")
ax.set_ylabel("Causa (GHE)")
ax.invert_yaxis()
plt.tight_layout()
plt.show()
```

# **Ranking top/bottom países**

ranking de los 5 países con la esperanza de vida más alta y más baja en 2021,

```python
import pandas as pd

ruta = "RELAY_WHS.csv"
df = pd.read_csv(ruta)

# 1. Filtrar: indicador, año y sexo
fx = (
    (df["IND_NAME"] == "Life expectancy (at birth)") &
    (df["DIM_TIME"] == 2021) &
    (df["DIM_SEX"]  == "TOTAL")
)
vida21 = df.loc[fx, ["GEO_NAME_SHORT", "AMOUNT_N"]].copy()

# 2. Ordenar de mayor a menor y de menor a mayor
top5    = vida21.sort_values("AMOUNT_N", ascending=False).head(5)
bottom5 = vida21.sort_values("AMOUNT_N", ascending=True ).head(5)

# 3. Mostrar resultados
print("Top 5 esperanza de vida 2021")
print(top5.to_string(index=False))

print("\nBottom 5 esperanza de vida 2021")
print(bottom5.to_string(index=False))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# --- 1 · Filtrar indicador, año y sexo ----------------------------------
mask = (
    (df["IND_NAME"] == "Life expectancy (at birth)") &
    (df["DIM_TIME"] == 2021) &
    (df["DIM_SEX"]  == "TOTAL")
)

vida21 = df.loc[mask, ["GEO_NAME_SHORT", "AMOUNT_N"]].copy()

# --- 2 · Top-5 y Bottom-5 ----------------------------------------------
top5    = vida21.sort_values("AMOUNT_N", ascending=False).head(5)
bottom5 = vida21.sort_values("AMOUNT_N", ascending=True ).head(5)

print("Top 5 esperanza de vida 2021")
print(top5.to_string(index=False))

print("\nBottom 5 esperanza de vida 2021")
print(bottom5.to_string(index=False))

# --- 3 · Gráfico comparativo (opcional, pero útil) ----------------------
fig, axes = plt.subplots(1, 2, figsize=(12, 5), sharex=False)

# Top-5 barras horizontales
axes[0].barh(top5["GEO_NAME_SHORT"], top5["AMOUNT_N"])
axes[0].set_title("Top 5")
axes[0].invert_yaxis()  # mayor arriba

# Bottom-5 barras horizontales
axes[1].barh(bottom5["GEO_NAME_SHORT"], bottom5["AMOUNT_N"])
axes[1].set_title("Bottom 5")
axes[1].invert_yaxis()

for ax in axes:
    ax.set_xlabel("Años de vida")
    ax.grid(alpha=0.3)

plt.suptitle("Esperanza de vida al nacer · 2021 (TOTAL)")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# --- 1 · Filtrar indicador, año y sexo -------------------------------
mask = (
    (df["IND_NAME"] == "Life expectancy (at birth)") &
    (df["DIM_TIME"] == 2021) &
    (df["DIM_SEX"]  == "TOTAL")
)

vida21 = df.loc[mask, ["GEO_NAME_SHORT", "AMOUNT_N"]].copy()

# --- 2 · Top-5 y Bottom-5 -------------------------------------------
top5    = vida21.sort_values("AMOUNT_N", ascending=False).head(5)
bottom5 = vida21.sort_values("AMOUNT_N", ascending=True ).head(5)

# --- 3 · Gráfico con etiquetas numéricas ----------------------------
fig, axes = plt.subplots(1, 2, figsize=(12, 5), sharex=False)

# --- Top-5
bars_top = axes[0].barh(top5["GEO_NAME_SHORT"], top5["AMOUNT_N"])
axes[0].set_title("Top 5")
axes[0].invert_yaxis()
axes[0].set_xlabel("Años de vida")
axes[0].grid(alpha=0.3)
axes[0].bar_label(bars_top, fmt='%.1f', padding=3)  # ← números

# --- Bottom-5
bars_bot = axes[1].barh(bottom5["GEO_NAME_SHORT"], bottom5["AMOUNT_N"])
axes[1].set_title("Bottom 5")
axes[1].invert_yaxis()
axes[1].set_xlabel("Años de vida")
axes[1].grid(alpha=0.3)
axes[1].bar_label(bars_bot, fmt='%.1f', padding=3)  # ← números

plt.suptitle("Esperanza de vida al nacer · 2021 (TOTAL)")
plt.tight_layout()
plt.show()
```
