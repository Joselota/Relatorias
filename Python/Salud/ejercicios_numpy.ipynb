
# 🧮 Ejercicios con NumPy

Este notebook contiene ejercicios simples y aplicados para entender cómo usar `numpy` en Python.  
Vamos a trabajar con vectores, matrices y algunas operaciones comunes que luego usarás al preparar datos para Machine Learning.


```

# Importar la librería
import numpy as np
```

## 🟢 Ejercicio 1: Crear arrays
Crea un array de los números del 1 al 10.

```
arr = np.arange(1, 11)
print(arr)
```

## 🟢 Ejercicio 2: Operaciones elementales
Multiplica todos los elementos del array por 2.

```
arr_doble = arr * 2
print(arr_doble)
```

## 🟢 Ejercicio 3: Estadísticas básicas
Calcula la media, desviación estándar y máximo del array.

```

print("Media:", np.mean(arr))
print("Desviación estándar:", np.std(arr))
print("Máximo:", np.max(arr))
```

## 🟢 Ejercicio 4: Crear una matriz aleatoria
Crea una matriz 3x3 con números aleatorios entre 0 y 1.

```
matriz = np.random.rand(3, 3)
print(matriz)
```

## 🟢 Ejercicio 5: Indexación y slicing
Extrae la segunda fila de la matriz anterior.

```
segunda_fila = matriz[1, :]
print(segunda_fila)
```

## 🟢 Ejercicio 6: Aplicado a ML
Convierte una lista de listas en un array y escala dividiendo por 100.

```

valores = [[100, 200, 300], [400, 500, 600]]
np_array = np.array(valores)
np_scaled = np_array / 100
print(np_scaled)
```
