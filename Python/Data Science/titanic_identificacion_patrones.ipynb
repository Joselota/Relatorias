# 🔎 Identificación de Patrones - Dataset del Titanic
En este notebook analizaremos patrones relevantes en el dataset del Titanic a partir de variables categóricas y numéricas.

Se explorarán:
- Patrones de supervivencia por sexo, clase y edad
- Relación entre tarifa pagada y supervivencia
- Asociación entre variables numéricas

## 📥 1. Cargar el dataset
Asegúrate de subir `Titanic-Dataset.csv` al entorno de ejecución.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('Titanic-Dataset.csv')
df.head()
```

## 👥 2. Supervivencia por sexo y clase

```python
sns.countplot(data=df, x='Sex', hue='Survived')
plt.title('Supervivencia por Sexo')
plt.show()

sns.countplot(data=df, x='Pclass', hue='Survived')
plt.title('Supervivencia por Clase')
plt.show()
```

## 🎯 3. Distribución de edad y supervivencia

```python
sns.histplot(data=df, x='Age', hue='Survived', bins=30, kde=True)
plt.title('Distribución de Edad según Supervivencia')
plt.show()
```

## 💸 4. Relación entre tarifa (`Fare`) y supervivencia

```python
sns.boxplot(data=df, x='Survived', y='Fare')
plt.title('Distribución de Tarifas según Supervivencia')
plt.show()
```

## 🔗 5. Correlaciones numéricas

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='YlGnBu')
plt.title('Matriz de Correlaciones')
plt.show()
```

## ✅ Conclusiones
- Mujeres y pasajeros de primera clase tuvieron mayor supervivencia.
- La edad y la tarifa pagada muestran patrones diferenciados.
- El análisis visual permite inferir relaciones clave antes de modelar.
