# 🧠 Estructuras de Datos Básicas en Python
Este notebook te guiará por las estructuras de datos fundamentales en Python: variables simples, listas, tuplas, conjuntos y diccionarios.

---

## 1. Variables simples
Los tipos de datos más básicos en Python:

```python
nombre = "Ana"
edad = 30
temperatura = 36.6
vacunado = True
print(nombre, edad, temperatura, vacunado)
```

## 2. Listas (`list`)
Colección ordenada y modificable:

```python
frutas = ["manzana", "pera", "naranja"]
print(frutas[0])  # Acceder al primer elemento
frutas.append("plátano")  # Agregar
print(frutas)
```

## 3. Tuplas (`tuple`)
Colección ordenada e inmutable:

```python
coordenadas = (40.7128, -74.0060)
print(coordenadas)
```

## 4. Conjuntos (`set`)
Colección no ordenada y sin elementos duplicados:

```python
turnos = {"mañana", "tarde", "noche", "mañana"}
print(turnos)
```

## 5. Diccionarios (`dict`)
Colección de pares clave-valor:

```python
paciente = {
    "nombre": "Ana",
    "edad": 30,
    "vacunado": True
}
print(paciente["nombre"])
paciente["enfermedad"] = "Hipertensión"
print(paciente)
```
