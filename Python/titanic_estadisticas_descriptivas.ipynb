# 📊 Estadísticas Descriptivas con el Dataset del Titanic
Este notebook presenta cómo aplicar estadísticas descriptivas para comprender mejor los datos numéricos del dataset del Titanic.

## 📥 1. Cargar el dataset
Asegúrate de subir `Titanic-Dataset.csv` a tu entorno de ejecución.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carga del dataset
df = pd.read_csv('Titanic-Dataset.csv')
df.head()
```

## 📊 2. Estadísticas básicas de resumen

```python
# Estadísticas generales de columnas numéricas
df.describe()
```

```python
# Estadísticas para columnas categóricas
df.describe(include='object')
```

## 🎯 3. Medidas de tendencia central y dispersión

```python
print("Media de edad:", df['Age'].mean())
print("Mediana de edad:", df['Age'].median())
print("Moda de edad:", df['Age'].mode()[0])

print("Mínimo:", df['Fare'].min())
print("Máximo:", df['Fare'].max())
print("Desviación estándar:", df['Fare'].std())
```

## 📈 4. Visualizaciones para entender la distribución

```python
# Histograma de edad
sns.histplot(df['Age'].dropna(), kde=True)
plt.title('Distribución de Edad')
plt.show()
```

```python
# Boxplot de tarifa (Fare)
sns.boxplot(x=df['Fare'])
plt.title('Boxplot de Tarifa')
plt.show()
```

## ✅ Conclusión
- Las estadísticas descriptivas nos permiten tener una visión inicial del comportamiento de los datos.
- Complementar con gráficos ayuda a detectar patrones, outliers y distribuciones.
