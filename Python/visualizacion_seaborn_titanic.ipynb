# 🌊 Visualización de Datos con Seaborn (Titanic Dataset)
Este notebook introduce visualizaciones con la librería `seaborn` usando el dataset del Titanic.
- Gráficos categóricos y numéricos
- Distribuciones y relaciones
- Personalización de visualizaciones

## 📥 1. Cargar el dataset
Sube el archivo `Titanic-Dataset.csv` antes de ejecutar el notebook.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('Titanic-Dataset.csv')
df.head()
```

## 📊 2. Countplot - Supervivencia por clase y sexo

```python
sns.countplot(data=df, x='Pclass', hue='Sex')
plt.title('Distribución de Pasajeros por Clase y Sexo')
plt.show()
```

## 📈 3. Histplot - Distribución de edades por supervivencia

```python
sns.histplot(data=df, x='Age', hue='Survived', bins=30, kde=True)
plt.title('Distribución de Edad por Supervivencia')
plt.show()
```

## 📉 4. Boxplot - Tarifa por clase y sexo

```python
sns.boxplot(data=df, x='Pclass', y='Fare', hue='Sex')
plt.title('Distribución de Tarifa por Clase y Sexo')
plt.show()
```

## 🔗 5. Heatmap - Correlaciones entre variables numéricas

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Matriz de Correlaciones')
plt.show()
```
