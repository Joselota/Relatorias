```python
df['PERIODO']= pd.to_datetime(df['PERIODO'], dayfirst=True)
```

# **Gráficos**

```python
import matplotlib.pyplot as plt
import seaborn as sns

df["REGIÓN"].value_counts().plot(kind="barh", figsize=(8,6))
plt.title("Inscritos por región")
plt.show()


```

```python
df['REGIÓN'] = df['REGIÓN'].str.replace("¿","'",regex=True)
```

```python
orden_tramos = ['0 a 19 años', '40 a 59 años', '20 a 39 años', '60 a 79 años', '80 a más años', 'Sin información']

sns.countplot(data=df, x='EDAD_TRAMO', order=orden_tramos)
plt.title("Distribución por edad")
plt.ylabel("Cantidad")
plt.xlabel("Tramo de edad")
plt.xticks(rotation=45)
plt.show()
```

```python
total_inscritos_sum = df['TOTAL_INSCRITOS'].sum()
print(f"The total sum of 'TOTAL_INSCRITOS' is: {total_inscritos_sum}")
```

```python
df_inscritos_por_edad = df.groupby('EDAD_TRAMO')['TOTAL_INSCRITOS'].sum().reset_index()

# Reorder the DataFrame based on the custom 'orden_tramos' for better visualization
df_inscritos_por_edad['EDAD_TRAMO'] = pd.Categorical(df_inscritos_por_edad['EDAD_TRAMO'], categories=orden_tramos, ordered=True)
df_inscritos_por_edad = df_inscritos_por_edad.sort_values('EDAD_TRAMO')

display(df_inscritos_por_edad)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
bars = sns.barplot(x='EDAD_TRAMO', y='TOTAL_INSCRITOS', data=df_inscritos_por_edad, palette='Blues', hue='EDAD_TRAMO', legend=False)
plt.title('Total Registered Individuals by Age Demographic')
plt.xlabel('Age Demographic')
plt.ylabel('Total Registered Individuals')
plt.xticks(rotation=45, ha='right')

# Add the values on top of the bars
for bar in bars.patches:
    bars.annotate(format(bar.get_height(), '.0f'),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   size=10, xytext=(0, 8),
                   textcoords='offset points')

plt.tight_layout()
plt.show()
```

```python
tabla = pd.crosstab(df['REGIÓN'],df['EDAD_TRAMO'], values=df['TOTAL_INSCRITOS'], aggfunc='sum')
tabla
```

```python
sns.heatmap(tabla, annot=False, cmap='YlGnBu')
plt.title('Distribución de tramos de edad por región')
plt.xlabel('Tramo de edad')
plt.ylabel('Región')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Group data by SEXO and EDAD_TRAMO and sum TOTAL_INSCRITOS
df_sexo_edad = df.groupby(['SEXO', 'EDAD_TRAMO'])['TOTAL_INSCRITOS'].sum().reset_index()

# Ensure EDAD_TRAMO is ordered correctly for plotting
df_sexo_edad['EDAD_TRAMO'] = pd.Categorical(df_sexo_edad['EDAD_TRAMO'], categories=orden_tramos, ordered=True)
df_sexo_edad = df_sexo_edad.sort_values('EDAD_TRAMO')

plt.figure(figsize=(12, 7))
bars = sns.barplot(x='EDAD_TRAMO', y='TOTAL_INSCRITOS', hue='SEXO', data=df_sexo_edad, palette='YlGnBu')
plt.title('Total Registered Individuals by Age Demographic and Sex')
plt.xlabel('Age Demographic')
plt.ylabel('Total Registered Individuals')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Sex')

# Add the values on top of the bars
for bar in bars.patches:
    bars.annotate(format(bar.get_height(), '.0f'),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   size=8, xytext=(0, 5),
                   textcoords='offset points')

plt.tight_layout()
plt.show()
```
