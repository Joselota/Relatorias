# Ejemplos de Estructura de datos: Control de Flujo en Python​

* Usar un if para clasificar temperaturas como “normal”, “febrícula” o “fiebre”.
* Recorrer una lista de pacientes con un for e imprimir sus edades.
* Crear un ciclo while que imprima turnos hasta que llegue a un número límite.

```python
# Usar un if para clasificar temperaturas como “normal”, “febrícula” o “fiebre”.

temp = float(input("Ingrese la temperatura: "))

if temp <= 36:
  print("Normal")
elif temp < 38:
  print("Febrícula")
else:
  print("Fiebre")
```

```python
#Recorrer una lista de pacientes con un for e imprimir sus edades.
pacientes = [
    {"nombre": "Ana", "edad": 34,"lugar":"Stgo"},
    {"nombre": "Luis", "edad": 28},
    {"nombre": "María", "edad": 45},
    {"nombre": "Jorge", "edad": 52},
    {"nombre": "Sofía", "edad": 22},
    {"nombre": "Carlos", "edad": 37},
    {"nombre": "Laura", "edad": 49}
]

for paciente in pacientes:
    print(paciente["nombre"], "-" ,paciente["edad"], "años")
```

```python
# Crear un ciclo while que imprima turnos hasta que llegue a un número límite.
limite = 5
turno = 0

while turno <= limite:
    print(f"Turno número {turno}")
    turno = turno + 1
```
