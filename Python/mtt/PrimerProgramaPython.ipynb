Script de CHAT GPT

```python
print("Bienvenido a la mini calculadora de Python!")

num1 = float(input("Ingresa el primer número: "))

# Pedimos num2 hasta que sea un número distinto de 0
while True:
    try:
        num2 = float(input("Ingresa el segundo número (≠ 0): "))
        if num2 == 0:
            print("⛔ El segundo número no puede ser 0. Intenta de nuevo.")
            continue
        break
    except ValueError:
        print("⚠️ Ingresa un número válido (por ejemplo, 12 o 12.5).")

print("La suma es:", num1 + num2)
print("La resta es:", num1 - num2)
print("La multiplicación es:", num1 * num2)
print("La división es:", num1 / num2)
```

**3 soluciones que nos entrego CHAT GPT**

```python
# 1) Con el “operador morsa” (:=) en el while
print("Bienvenido a la mini calculadora de Python!")

num1 = float(input("Ingresa el primer número: "))

while (num2 := float(input("Ingresa el segundo número (≠ 0): "))) == 0:
    print("⛔ No puede ser 0. Intenta de nuevo.")

print("La suma es:", num1 + num2)
print("La resta es:", num1 - num2)
print("La multiplicación es:", num1 * num2)
print("La división es:", num1 / num2)
```

```python
#2) Con un do-while “manual” usando while True
print("Bienvenido a la mini calculadora de Python!")

num1 = float(input("Ingresa el primer número: "))

while True:
    num2 = float(input("Ingresa el segundo número (≠ 0): "))
    if num2 != 0:
        break
    print("⛔ No puede ser 0. Intenta de nuevo.")

print("La suma es:", num1 + num2)
print("La resta es:", num1 - num2)
print("La multiplicación es:", num1 * num2)
print("La división es:", num1 / num2)
```

```python
#3) Validando “a posteriori” con try/except y reintento
print("Bienvenido a la mini calculadora de Python!")

num1 = float(input("Ingresa el primer número: "))

while True:
    num2 = float(input("Ingresa el segundo número: "))
    try:
        resultado_div = num1 / num2   # si num2 == 0, cae al except
        break
    except ZeroDivisionError:
        print("⛔ El segundo número no puede ser 0. Intenta de nuevo.")

print("La suma es:", num1 + num2)
print("La resta es:", num1 - num2)
print("La multiplicación es:", num1 * num2)
print("La división es:", resultado_div)
```

**Nuevo ejemplo**

```python
# usuario ingresa valor de sueldo bruto
while True:
    try:
        sueldo_bruto = float(input("Ingresa tu sueldo bruto: "))
        break
    except ValueError:
        print("⛔ Entrada inválida. Por favor, ingresa un número para el sueldo bruto.")

# Se aplica el descuento
descuento = sueldo_bruto * 0.1   # 10% de descuento

# Calcula sueldo liquido
sueldo_liquido = sueldo_bruto - descuento

# Muestra en pantalla sueldo líquido
print("Tu sueldo líquido es:", sueldo_liquido)
```
