**Función promedio que usa lista**

```python
def promedio(notas):
 return sum(notas) / len(notas)

notas = [5.5, 6.0, 6.3, 6.8]

print("Promedio:", round(promedio(notas), 2))
```

```python
import csv

total = 0

with open("ventas.csv", "r") as archivo:
    lector = csv.reader(archivo)
    next(lector)  # omite encabezado​
    for fila in lector:
        total = total + int(fila[2])
        print(fila)

print("Total de ventas:", total)
```

```python
import csv

total = 0

with open("ventas.csv", "r") as archivo:
    lector = csv.reader(archivo)
    for indice, colum in enumerate(lector):
        if indice == 0:
            continue  # Salta la primera fila
        total = total + int(colum[2])
        print(colum)

print("Total de ventas:", total)
```

```python
import csv
ventas = [ ["fecha",      "producto", "monto"],
           ["2025-01-10", "pan",         1200],
           ["2025-01-09", "leche",       1500],
           ["2025-01-12", "huevos",      1700] ]

with open("ventas_nuevas.csv", "w", newline="") as f:
      escritor = csv.writer(f)
      escritor.writerows(ventas)
```

```python
import csv

with open("ventas.csv", "r") as f:
    lector = csv.DictReader(f)
    for fila in lector:
        #print(fila["fecha"], "→", fila["producto"], "→", fila["monto"])
        print(fila["producto"], "→", fila["monto"])
```

```python
import csv

resumen = {}

with open("ventas.csv", "r") as f:
    lector = csv.DictReader(f)
    for fila in lector:
     producto = fila["producto"]
     monto = int(fila["monto"])
     resumen[producto] = resumen.get(producto, 0) + monto
     print(resumen)

#print(resumen)
```

```python
import csv

print(resumen)

with open("resumen_ventas.csv", "w", newline="") as f:
    escritor = csv.writer(f)
    escritor.writerow(["producto", "total"])
    for producto, total in resumen.items():
        print(producto, "→", total)
        escritor.writerow([producto, total])

```

```python
import csv

ventas_altas = []

with open("ventas.csv", "r") as f:
    lector = csv.DictReader(f)
    for col in lector:
        monto = int(col["monto"])
        if monto > 1500:
            ventas_altas.append(col)

print("Ventas altas:", ventas_altas)
```
