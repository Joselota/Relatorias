```python
import pandas as pd

df_metro = pd.read_excel('metro_20251119-oficio-4770_2013.xlsx', sheet_name='Abiertos', skiprows=7)
df_metro.head()
```

```python
cols_estaciones = ["CODIGO","NOMBRE FANTASIA","DIRECCION","COMUNA",'LONGITUD', 'LATITUD']
cols_estaciones = [c for c in cols_estaciones if c in df_metro.columns]
```

```python
estaciones = (
    df_metro[cols_estaciones]
    .drop_duplicates(subset=["CODIGO", "NOMBRE FANTASIA", "COMUNA"])
    .sort_values(["COMUNA", "NOMBRE FANTASIA"])
)
```

```python
estaciones.head(10)
```

```python
estaciones.to_csv("estaciones_metro.csv", index=False)
```

```python
import folium

center_lat = estaciones.LATITUD.mean()
center_lon = estaciones.LONGITUD.mean()

m = folium.Map(location=[center_lat,center_lon], width="80%", height="80%")

for _, row in estaciones.iterrows():
  folium.CircleMarker(
      location=[row["LATITUD"],row["LONGITUD"]],
      radius=2,
      popup=f"{row['NOMBRE FANTASIA']} ({row['COMUNA']})",
  ).add_to(m)

m

```

```python
colores = {
    "L1": "red",
    "L2": "orange",
    "L3": "brown",
    "L4": "#2A2576",
    "L4A": "#0171BB",
    "L5": "darkgreen",
    "L6": "purple"
}
```

```python
import folium

center_lat = estaciones["LATITUD"].mean()
center_lon = estaciones["LONGITUD"].mean()

m = folium.Map(
    location=[center_lat, center_lon],
    zoom_start=11,
    width="100%",
    height="80%"
)
```

```python
for _, row in estaciones.dropna(subset=["LATITUD", "LONGITUD"]).iterrows():
    linea = row["CODIGO"]
    color = colores.get(linea, "gray")  # gris si es una línea no mapeada

    folium.CircleMarker(
        location=[row["LATITUD"], row["LONGITUD"]],
        radius=5,
        color=color,
        fill=True,
        fill_opacity=0.8,
        popup=f"{row['NOMBRE FANTASIA']} (Línea {linea})"
    ).add_to(m)

m
```

```python
legend_html = ''
legend_html = '''<div style="
                      position: fixed;
                      top: 100px;
                      left: 50px;
                      width: 150px;
                      height: auto;
                      border:2px solid grey;
                      z-index:9999;
                      font-size:12px;
                      background-color:white;
                      padding:10px;
                  ">
                  <p><b>Leyenda Líneas de Metro</b></p>
                  '''

for line, color in colores.items():
    legend_html += f'<i style="background:{color}; width:10px; height:10px; display:inline-block;"></i> {line}<br>'

legend_html += '</div>'

m.get_root().html.add_child(folium.Element(legend_html))

m
```

```python
# Al final de la construcción del mapa
m.save("mapa_metro_santiago.html")
```
