```python
import matplotlib.pyplot as plt
```

**Gráfico de línea**

```python
plt.figure(figsize=(10,4))
plt.plot(df_arroz["fecha"], df_arroz["monto"])
plt.title("Ventas de arroz en el tiempo")
plt.xlabel("Fecha")
plt.ylabel("Monto")
plt.grid(True, linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

**Gráfico de barras**

```python
df_arroz.head()
```

```python
df_arroz["mes"] = df_arroz["fecha"].dt.to_period("M")
```

```python
df_arroz.head()
```

```python
# Ventas por mes
ventas_arroz_mes = df_arroz.groupby("mes")["monto"].sum()
```

```python
plt.figure(figsize=(8,4))
ventas_arroz_mes.plot(kind="bar", color="#37C5F0")
plt.title("Ventas mensuales de arroz")
plt.xlabel("Mes")
plt.ylabel("Monto total")
plt.grid(True, linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,4))
ventas_arroz_mes.plot(kind="barh", color="#37C5F0")
plt.title("Ventas mensuales de arroz")
plt.xlabel("Mes")
plt.ylabel("Monto total")
plt.grid(True, linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()
```
