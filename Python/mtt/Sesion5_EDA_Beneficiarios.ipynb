```python
df['BENEFICIARIOS'] = pd.to_numeric(df['BENEFICIARIOS'], errors='coerce')
```

```python
df['MES_INFORMACION'].head()
```

```python
df['AÑO'] = df['MES_INFORMACION'].astype(str).str.slice(0,4).astype(int)
df['MES'] = df['MES_INFORMACION'].astype(str).str.slice(4,6).astype(int)
```

```python
beneficiarios_por_tramo = df.groupby('TRAMO')['BENEFICIARIOS'].sum().reset_index()

display(beneficiarios_por_tramo)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
plt.pie(beneficiarios_por_tramo['BENEFICIARIOS'],
        labels=beneficiarios_por_tramo['TRAMO'],
        autopct='%1.1f%%',
        startangle=90,
        colors=['lightcoral', 'lightskyblue', 'lightgreen', 'gold', 'plum']) # Changed colors
plt.title('Distribución de Beneficiarios por Tramo')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
beneficiarios_por_region = df.groupby('REGIÓN')['BENEFICIARIOS'].sum().reset_index()

display(beneficiarios_por_region)
```

```python
# Define the regions to be excluded
excluded_regions = ['Desconocida', 'Example Region']

# Filter the DataFrame to exclude these regions
df_filtered = df[~df['REGIÓN'].isin(excluded_regions)]

# Group by 'REGIÓN' and sum 'BENEFICIARIOS' from the filtered DataFrame
beneficiarios_por_region = df_filtered.groupby('REGIÓN')['BENEFICIARIOS'].sum().reset_index()

display(beneficiarios_por_region)
```

```python
#beneficiarios_por_region

plt.figure(figsize=(8, 6))
beneficiarios_por_region.plot(kind="barh")
plt.title("Cantidad de registros por región")
plt.xlabel("Registros (combinaciones de variables)")
plt.ylabel("Región")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

import seaborn as sns


plt.figure(figsize=(10, 6))

ax = sns.barplot(x='REGIÓN', y='BENEFICIARIOS', data=beneficiarios_por_region, palette='viridis', hue='REGIÓN', legend=False)



# Add the sum of beneficiaries on top of each bar

for p in ax.patches:
  ax.annotate(f'{p.get_height():,}',
  (p.get_x() + p.get_width() / 2., p.get_height()),
  ha='center', va='center',
  xytext=(0, 9),
  textcoords='offset points')


plt.title('Beneficiarios por Region')
plt.xlabel('Región')
plt.ylabel('Total de Beneficiarios')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Sort the DataFrame by 'BENEFICIARIOS' in descending order
beneficiarios_por_region_sorted = beneficiarios_por_region.sort_values(by='BENEFICIARIOS', ascending=True)

plt.figure(figsize=(12, 8))
ax = sns.barplot(x='BENEFICIARIOS', y='REGIÓN', data=beneficiarios_por_region_sorted, palette='viridis', hue='REGIÓN', legend=False)

# Add the sum of beneficiaries next to each bar
for p in ax.patches:
    width = p.get_width()
    plt.text(width + 50000,
             p.get_y() + p.get_height() / 2,
             f'{width:,.0f}',
             ha='left', va='center')

plt.title('Beneficiarios por Región (Ordenado)')
plt.xlabel('Total de Beneficiarios')
plt.ylabel('Región')
plt.tight_layout()
plt.show()
```
