```python
# === Mapa 10K mejorado (GeoPandas + .explore) ===
import gpxpy, geopandas as gpd
from shapely.geometry import LineString, Point
import math

GPX_FILE = "activity_20802394121 (10K-Maipu).gpx"

with open(GPX_FILE, "r") as f:
    gpx = gpxpy.parse(f)

seg = gpx.tracks[0].segments[0]
pts = seg.points

# (lon, lat)
coords = [(p.longitude, p.latitude) for p in pts]

# Lnea principal
gdf_line = gpd.GeoDataFrame(
    {"name": ["Ruta 10K"]},
    geometry=[LineString(coords)],
    crs="EPSG:4326"
)

# Inicio y fin
gdf_start = gpd.GeoDataFrame(
    {"label": ["Inicio"]},
    geometry=[Point(coords[0])],
    crs="EPSG:4326"
)

gdf_end = gpd.GeoDataFrame(
    {"label": ["Fin"]},
    geometry=[Point(coords[-1])],
    crs="EPSG:4326"
)

# Funcin para distancia (para calcular km)
def hav(lat1, lon1, lat2, lon2):
    R = 6371
    dlat = math.radians(lat2-lat1)
    dlon = math.radians(lon2-lon1)
    a = math.sin(dlat/2)**2 + math.cos(math.radians(lat1))*math.cos(math.radians(lat2))*math.sin(dlon/2)**2
    return R * (2 * math.atan2(math.sqrt(a), math.sqrt(1-a)))

# Crear puntos cada 1 km aprox.
km_list = []
acc = 0
k = 0
last = coords[0]
for lon, lat in coords[1:]:
    d = hav(last[1], last[0], lat, lon)
    acc += d
    if acc >= 1:
        k += 1
        km_list.append({"km": k, "geometry": Point(lon, lat)})
        acc = 0
    last = (lon, lat)

gdf_km = gpd.GeoDataFrame(km_list, crs="EPSG:4326")

# === MAPA ===
m = gdf_line.explore(
    tiles="CartoDB Positron",        # Fondo ms limpio y moderno
    color="#ff4a3d",                 # Rojo suave bonito
    style_kwds={"weight": 5},        # Lnea ms contundente
    tooltip=False,
    name="Ruta 10K"
)

# Inicio / Fin
gdf_start.explore(
    m=m, marker_type="marker",
    color="green", tooltip=["label"], popup=["label"]
)
gdf_end.explore(
    m=m, marker_type="marker",
    color="red", tooltip=["label"], popup=["label"]
)

# Marcadores km
if len(gdf_km) > 0:
    gdf_km["texto"] = gdf_km["km"].astype(str) + " km"
    gdf_km.explore(m=m, marker_type="circle", color="blue",
                   marker_kwds={"radius": 5}, tooltip=["texto"])

# Ajustar encuadre exacto a la ruta
minx, miny, maxx, maxy = gdf_line.total_bounds
m.fit_bounds([[miny, minx], [maxy, maxx]])

m  # o m.save("10k_mejorado.html")
```

```python
# === Colorear la ruta por PACE (min/km) con GeoPandas ===
import math
import pandas as pd
import geopandas as gpd
from shapely.geometry import LineString, Point

def hav_km(lat1, lon1, lat2, lon2):
    R = 6371.0
    dlat = math.radians(lat2 - lat1)
    dlon = math.radians(lon2 - lon1)
    a = math.sin(dlat/2)**2 + math.cos(math.radians(lat1))*math.cos(math.radians(lat2))*math.sin(dlon/2)**2
    return R * (2 * math.atan2(math.sqrt(a), math.sqrt(1-a)))

# 1) Tomamos puntos del GPX ya cargado en variables `seg` o `pts`
pts = seg.points  # si usas otro nombre, ajstalo

# 2) Construir segmentos consecutivos con distancia y delta tiempo
rows = []
for i in range(len(pts)-1):
    p1, p2 = pts[i], pts[i+1]
    # algunos GPX no traen tiempo, valida antes:
    if (p1.time is None) or (p2.time is None):
        continue
    dt = (p2.time - p1.time).total_seconds()
    if dt <= 0:
        continue

    d_km = hav_km(p1.latitude, p1.longitude, p2.latitude, p2.longitude)
    if d_km == 0:
        continue

    pace_min_km = (dt / 60.0) / max(d_km, 1e-9)   # min por km
    geom = LineString([(p1.longitude, p1.latitude), (p2.longitude, p2.latitude)])
    rows.append({
        "t_ini": pd.to_datetime(p1.time),
        "t_fin": pd.to_datetime(p2.time),
        "dist_km": d_km,
        "dt_s": dt,
        "pace_min_km": pace_min_km,
        "geometry": geom
    })

gdf_seg = gpd.GeoDataFrame(rows, crs="EPSG:4326")
assert len(gdf_seg) > 0, "No se pudieron construir segmentos con tiempo vlido."

# 3) Limpieza rpida de outliers (opcional pero til)
#    recorta pace a percentiles 2-98 para evitar picos raros por GPS
p2, p98 = gdf_seg["pace_min_km"].quantile([0.02, 0.98]).tolist()
gdf_seg["pace_plot"] = gdf_seg["pace_min_km"].clip(lower=p2, upper=p98)

# Quitar o convertir timestamps antes de explorar
gdf_seg["t_ini"] = gdf_seg["t_ini"].astype(str)
gdf_seg["t_fin"] = gdf_seg["t_fin"].astype(str)

# longitud objetivo por tramo (en km)  0.03 = ~30 metros
TARGET_DIST = 0.03

grouped = []
acc_dist = 0
current_coords = []
current_paces = []

for i, row in gdf_seg.iterrows():
    line = row.geometry
    (lon1, lat1), (lon2, lat2) = line.coords
    dist = row.dist_km
    pace = row.pace_plot

    acc_dist += dist
    current_coords.extend([(lon1, lat1), (lon2, lat2)])
    current_paces.append(pace)

    # si acumulamos ms que el tramo deseado  formamos segmento
    if acc_dist >= TARGET_DIST:
        grouped.append({
            "avg_pace": sum(current_paces) / len(current_paces),
            "geometry": LineString(current_coords)
        })
        acc_dist = 0
        current_coords = []
        current_paces = []

# construir GeoDataFrame limpio
gdf_smooth = gpd.GeoDataFrame(grouped, crs="EPSG:4326")

m = gdf_smooth.explore(
    tiles="CartoDB Positron",
    column="avg_pace",
    cmap="plasma",         # o "viridis", "turbo"
    scheme="quantiles",
    k=6,
    legend=True,
    style_kwds={"weight": 7, "opacity": 0.95},
    name="Ritmo suavizado"
)


# Marcar inicio / fin (si no lo habas agregado an)
start_pt.explore(m=m, marker_type="marker", color="green", tooltip=["label"])
end_pt.explore(m=m, marker_type="marker", color="red", tooltip=["label"])

minx, miny, maxx, maxy = gdf_smooth.total_bounds
m.fit_bounds([[miny, minx], [maxy, maxx]])

m


```

```python
import numpy as np
import pandas as pd
import geopandas as gpd
from shapely.geometry import Polygon
from shapely.geometry import Point


# --- funcin para grilla y promedio ponderado por longitud dentro de la celda ---
def grid_heatmap_along_route_fixedbins(gdf_smooth, grid_m=150, buffer_m=60, bins=None, labels=None):
    gdfm = gdf_smooth.to_crs(32719)

    # corredor alrededor de la ruta (para recortar la grilla)
    buffer = gdfm.buffer(buffer_m).unary_union

    # grilla rectangular sobre bbox de la ruta
    minx, miny, maxx, maxy = gdfm.total_bounds
    xs = np.arange(minx, maxx + grid_m, grid_m)
    ys = np.arange(miny, maxy + grid_m, grid_m)
    polys = [Polygon([(x,y),(x+grid_m,y),(x+grid_m,y+grid_m),(x,y+grid_m)])
             for x in xs[:-1] for y in ys[:-1]]
    grid = gpd.GeoDataFrame(geometry=polys, crs=gdfm.crs)
    grid["cell_id"] = grid.index

    # solo celdas que tocan el corredor
    grid_cut = grid[grid.intersects(buffer)]

    # interseccin ruta-celda
    inter = gpd.overlay(
        gdfm[["avg_pace", "geometry"]],
        grid_cut[["cell_id", "geometry"]],
        how="intersection"
    )
    if inter.empty:
        return gpd.GeoDataFrame(columns=["geometry","pace_cell","pace_band"], crs=4326)

    inter["len_m"] = inter.length

    # promedio ponderado por longitud SIN apply (adis warning)
    tmp = inter[["cell_id", "avg_pace", "len_m"]].copy()
    tmp["aw"] = tmp["avg_pace"] * tmp["len_m"]
    
    agg = (
        tmp.groupby("cell_id", as_index=False)
           .agg(sum_aw=("aw", "sum"), sum_w=("len_m", "sum"))
    )
    agg["pace_cell"] = agg["sum_aw"] / agg["sum_w"]
    agg = agg[["cell_id", "pace_cell"]]

    grid_pace = grid_cut.merge(agg, on="cell_id", how="left").to_crs(4326)

    # bins y etiquetas humanas
    if bins is None:
        bins = [0, 5.5, 6.0, 6.5, 7.0, 8.0, 99]   # min/km
    if labels is None:
        labels = ["<5:30", "5:305:59", "6:006:29", "6:306:59", "7:007:59", "8:00"]

    grid_pace["pace_band"] = pd.cut(grid_pace["pace_cell"], bins=bins, labels=labels, include_lowest=True)
    return grid_pace


def add_km_markers(gdf_smooth, m):
    # proyectar para medir distancia acumulada
    gdfm = gdf_smooth.to_crs(32719)
    gdfm["segment_len_m"] = gdfm.length
    gdfm["dist_acum_m"] = gdfm["segment_len_m"].cumsum()

    # puntos cada 1000m
    kms = np.arange(1000, gdfm["dist_acum_m"].max(), 1000)
    marks = []
    for k in kms:
        row = gdfm.iloc[(gdfm["dist_acum_m"] - k).abs().argmin()]
        # tomar punto medio del segmento
        x, y = list(row.geometry.coords)[1]
        marks.append((k/1000, Point(row.geometry.centroid.x, row.geometry.centroid.y)))

    gdf_marks = gpd.GeoDataFrame(
        {"km": [km for km, _ in marks]},
        geometry=[pt for _, pt in marks],
        crs=gdfm.crs
    ).to_crs(4326)

    gdf_marks.explore(
        m=m, marker_type="marker",
        popup=["km"], tooltip=["km"],
        marker_kwds={"radius": 6, "color": "black", "fillColor": "white", "fillOpacity": 1}
    )
    return m



# --- construir heatmap limpio con bandas elegantes ---
grid_pace = grid_heatmap_along_route_fixedbins(
    gdf_smooth,
    grid_m=150,   # tamao de celda
    buffer_m=60   # ancho del corredor
)

# quitar celdas sin datos reales
grid_pace_clean = grid_pace.dropna(subset=["pace_cell"]).copy()

# mapa
m2 = grid_pace_clean.explore(
    tiles="CartoDB Positron",
    column="pace_band",            # categrico con etiquetas
    cmap="plasma",
    legend=True,
    style_kwds={"weight": 0.1, "fillOpacity": 0.55},
    name="Pace por zona (plasma)"
)

# superponer ruta y marcadores
gdf_smooth.explore(m=m2, color="#111", style_kwds={"weight": 4, "opacity": 0.95}, name="Ruta")
gdf_start.explore(m=m2, marker_type="marker", color="green")
gdf_end.explore(m=m2, marker_type="marker", color="red")

# encuadre
minx, miny, maxx, maxy = grid_pace_clean.total_bounds
m2.fit_bounds([[miny, minx], [maxy, maxx]])

m2.save("10k_pace_heatmap_plasma.html")
# m2

m_final = add_km_markers(gdf_smooth, m2)
m_final.save("10k_pace_heatmap_plasma_km.html")
m_final

```

```python
import numpy as np
import pandas as pd
import geopandas as gpd
from shapely.geometry import Polygon, Point
import folium  #  necesario para CircleMarker y DivIcon


# --- funcin para grilla y promedio ponderado por longitud dentro de la celda ---
def grid_heatmap_along_route_fixedbins(gdf_smooth, grid_m=150, buffer_m=60, bins=None, labels=None):
    gdfm = gdf_smooth.to_crs(32719)

    # corredor alrededor de la ruta (para recortar la grilla)
    buffer = gdfm.buffer(buffer_m).unary_union

    # grilla rectangular sobre bbox de la ruta
    minx, miny, maxx, maxy = gdfm.total_bounds
    xs = np.arange(minx, maxx + grid_m, grid_m)
    ys = np.arange(miny, maxy + grid_m, grid_m)
    polys = [Polygon([(x,y),(x+grid_m,y),(x+grid_m,y+grid_m),(x,y+grid_m)])
             for x in xs[:-1] for y in ys[:-1]]
    grid = gpd.GeoDataFrame(geometry=polys, crs=gdfm.crs)
    grid["cell_id"] = grid.index

    # solo celdas que tocan el corredor
    grid_cut = grid[grid.intersects(buffer)]

    # interseccin ruta-celda
    inter = gpd.overlay(
        gdfm[["avg_pace", "geometry"]],
        grid_cut[["cell_id", "geometry"]],
        how="intersection"
    )
    if inter.empty:
        return gpd.GeoDataFrame(columns=["geometry","pace_cell","pace_band"], crs=4326)

    inter["len_m"] = inter.length

    # promedio ponderado por longitud SIN apply (adis warning)
    tmp = inter[["cell_id", "avg_pace", "len_m"]].copy()
    tmp["aw"] = tmp["avg_pace"] * tmp["len_m"]
    agg = (
        tmp.groupby("cell_id", as_index=False)
           .agg(sum_aw=("aw", "sum"), sum_w=("len_m", "sum"))
    )
    agg["pace_cell"] = agg["sum_aw"] / agg["sum_w"]
    agg = agg[["cell_id", "pace_cell"]]

    grid_pace = grid_cut.merge(agg, on="cell_id", how="left").to_crs(4326)

    # bins y etiquetas humanas
    if bins is None:
        bins = [0, 5.5, 6.0, 6.5, 7.0, 8.0, 99]   # min/km
    if labels is None:
        labels = ["<5:30", "5:305:59", "6:006:29", "6:306:59", "7:007:59", "8:00"]

    grid_pace["pace_band"] = pd.cut(grid_pace["pace_cell"], bins=bins, labels=labels, include_lowest=True)
    return grid_pace


# --- marcadores elegantes por kilmetro (crculo + nmero grande) ---
def add_km_markers(gdf_smooth, m):
    # proyectar para medir distancia acumulada
    gdfm = gdf_smooth.to_crs(32719).copy()
    gdfm["segment_len_m"] = gdfm.length
    gdfm["dist_acum_m"] = gdfm["segment_len_m"].cumsum()

    # puntos cada 1000 m
    kms = np.arange(1000, gdfm["dist_acum_m"].max(), 1000)
    marks = []
    for k in kms:
        idx = (gdfm["dist_acum_m"] - k).abs().argmin()
        seg = gdfm.iloc[idx].geometry
        pt_m = seg.interpolate(0.5, normalized=True)  # punto medio del segmento
        marks.append((int(k/1000), pt_m))

    gdf_marks = gpd.GeoDataFrame(
        {"km": [km for km, _ in marks]},
        geometry=[pt for _, pt in marks],
        crs=gdfm.crs
    ).to_crs(4326)

    # dibujar: crculo blanco con borde negro + nmero encima
    for _, row in gdf_marks.iterrows():
        lat, lon = row.geometry.y, row.geometry.x

        folium.CircleMarker(
            location=[lat, lon],
            radius=12,
            weight=2,
            color="#111111",
            fill=True,
            fill_color="white",
            fill_opacity=0.95,
            popup=f"Km {int(row.km)}",
        ).add_to(m)

        folium.Marker(
            location=[lat, lon],
            icon=folium.DivIcon(
                html=f"""
                    <div style="
                        font-size:12px;
                        font-weight:700;
                        color:#111;
                        text-align:center;
                        transform: translate(-45%, -55%);
                    ">{int(row.km)}</div>
                """
            ),
        ).add_to(m)

    return m


# --- construir heatmap limpio con bandas elegantes ---
grid_pace = grid_heatmap_along_route_fixedbins(
    gdf_smooth,
    grid_m=150,   # tamao de celda
    buffer_m=60   # ancho del corredor
)

# quitar celdas sin datos reales
grid_pace_clean = grid_pace.dropna(subset=["pace_cell"]).copy()

# mapa
m2 = grid_pace_clean.explore(
    tiles="CartoDB Positron",
    column="pace_band",            # categrico con etiquetas
    cmap="plasma",
    legend=True,
    style_kwds={"weight": 0.1, "fillOpacity": 0.55},
    name="Pace por zona (plasma)"
)

# superponer ruta y marcadores de inicio/fin
gdf_smooth.explore(m=m2, color="#111", style_kwds={"weight": 4, "opacity": 0.95}, name="Ruta")
gdf_start.explore(m=m2, marker_type="marker", color="green")
gdf_end.explore(m=m2, marker_type="marker", color="red")

# encuadre
minx, miny, maxx, maxy = grid_pace_clean.total_bounds
m2.fit_bounds([[miny, minx], [maxy, maxx]])

m2.save("10k_pace_heatmap_plasma.html")

# aadir marcadores de km y exportar final
m_final = add_km_markers(gdf_smooth, m2)
m_final.save("10k_pace_heatmap_plasma_km.html")
m_final
```

```python
# === 10K Maip  GeoPandas + Folium (pace, heatmap, km markers con halo) ===

# ------------------ CONFIG ------------------
GPX_FILE   = "activity_20802394121 (10K-Maipu).gpx"  # cambia si corresponde
TILE       = "CartoDB Positron"
SMOOTH_M   = 30         # ~metros por tramo suavizado
GRID_M     = 150        # tamao de celda del heatmap (m)
BUFFER_M   = 60         # ancho del corredor alrededor de la ruta (m)
BINS_MINKM = [0, 5.5, 6.0, 6.5, 7.0, 8.0, 99]
BINS_LABEL = ["<5:30", "5:305:59", "6:006:29", "6:306:59", "7:007:59", "8:00"]

OUT_HTML_HEATMAP = "10k_pace_heatmap_plasma.html"
OUT_HTML_FINAL   = "10k_pace_heatmap_plasma_km.html"

# ------------------ IMPORTS ------------------
import math
import numpy as np
import pandas as pd
import gpxpy, geopandas as gpd
from shapely.geometry import LineString, Point, Polygon
import folium
from folium.plugins import Fullscreen, MeasureControl
from folium import LayerControl

# ------------------ UTILIDADES ------------------
def hav_km(lat1, lon1, lat2, lon2):
    R = 6371.0
    dlat = math.radians(lat2 - lat1)
    dlon = math.radians(lon2 - lon1)
    a = math.sin(dlat/2)**2 + math.cos(math.radians(lat1))*math.cos(math.radians(lat2))*math.sin(dlon/2)**2
    return R * (2 * math.atan2(math.sqrt(a), math.sqrt(1-a)))

def build_segments_with_pace(seg_points):
    """Segmentos consecutivos del GPX con pace (min/km) + limpieza outliers."""
    rows = []
    for i in range(len(seg_points) - 1):
        p1, p2 = seg_points[i], seg_points[i+1]
        if (p1.time is None) or (p2.time is None):
            continue
        dt = (p2.time - p1.time).total_seconds()
        if dt <= 0:
            continue
        d_km = hav_km(p1.latitude, p1.longitude, p2.latitude, p2.longitude)
        if d_km <= 0:
            continue
        pace_min_km = (dt / 60.0) / max(d_km, 1e-9)
        geom = LineString([(p1.longitude, p1.latitude), (p2.longitude, p2.latitude)])
        rows.append({
            "t_ini": pd.to_datetime(p1.time).strftime("%Y-%m-%d %H:%M:%S"),
            "t_fin": pd.to_datetime(p2.time).strftime("%Y-%m-%d %H:%M:%S"),
            "dist_km": d_km,
            "dt_s": dt,
            "pace_min_km": pace_min_km,
            "geometry": geom
        })
    gdf = gpd.GeoDataFrame(rows, crs="EPSG:4326")
    if gdf.empty:
        raise ValueError("No se pudieron construir segmentos (el GPX no trae timestamps?).")
    p2, p98 = gdf["pace_min_km"].quantile([0.02, 0.98]).tolist()
    gdf["pace_plot"] = gdf["pace_min_km"].clip(lower=p2, upper=p98)
    return gdf

def smooth_by_target_meters(gdf_seg, target_m=30):
    """Agrupa segmentos cortos en tramos ~target_m metros y promedia pace."""
    gdfm = gdf_seg.to_crs(32719)
    grouped, acc_m, coords, paces = [], 0.0, [], []
    for geom, pace in zip(gdfm.geometry, gdfm["pace_plot"]):
        (x1, y1), (x2, y2) = list(geom.coords)
        seg_len_m = math.dist((x1, y1), (x2, y2))
        acc_m += seg_len_m
        coords.extend([(x1, y1), (x2, y2)])
        paces.append(pace)
        if acc_m >= target_m:
            grouped.append({"avg_pace": float(np.mean(paces)), "geometry": LineString(coords)})
            acc_m, coords, paces = 0.0, [], []
    gdf_smooth_m = gpd.GeoDataFrame(grouped, crs=gdfm.crs)
    return gdf_smooth_m.to_crs(4326)

def grid_heatmap_along_route_fixedbins(gdf_smooth, grid_m=150, buffer_m=60, bins=None, labels=None):
    """Celdas recortadas al corredor de la ruta y pace ponderado por longitud."""
    gdfm = gdf_smooth.to_crs(32719)

    # corredor alrededor de la ruta
    buffer = gdfm.buffer(buffer_m).unary_union

    # grilla rectangular sobre bbox de la ruta
    minx, miny, maxx, maxy = gdfm.total_bounds
    xs = np.arange(minx, maxx + grid_m, grid_m)
    ys = np.arange(miny, maxy + grid_m, grid_m)
    polys = [Polygon([(x,y),(x+grid_m,y),(x+grid_m,y+grid_m),(x,y+grid_m)])
             for x in xs[:-1] for y in ys[:-1]]
    grid = gpd.GeoDataFrame(geometry=polys, crs=gdfm.crs)
    grid["cell_id"] = grid.index

    # solo celdas que tocan el corredor
    grid_cut = grid[grid.intersects(buffer)]

    # interseccin ruta-celda
    inter = gpd.overlay(
        gdfm[["avg_pace", "geometry"]],
        grid_cut[["cell_id", "geometry"]],
        how="intersection"
    )
    if inter.empty:
        return gpd.GeoDataFrame(columns=["geometry","pace_cell","pace_band"], crs=4326)

    inter["len_m"] = inter.length

    # promedio ponderado por longitud (sin .apply  sin warnings)
    tmp = inter[["cell_id", "avg_pace", "len_m"]].copy()
    tmp["aw"] = tmp["avg_pace"] * tmp["len_m"]
    agg = tmp.groupby("cell_id", as_index=False).agg(sum_aw=("aw","sum"), sum_w=("len_m","sum"))
    agg["pace_cell"] = agg["sum_aw"] / agg["sum_w"]
    agg = agg[["cell_id","pace_cell"]]

    grid_pace = grid_cut.merge(agg, on="cell_id", how="left").to_crs(4326)

    # bins y etiquetas humanas
    if bins is None:   bins = BINS_MINKM
    if labels is None: labels = BINS_LABEL
    grid_pace["pace_band"] = pd.cut(grid_pace["pace_cell"], bins=bins, labels=labels, include_lowest=True)
    return grid_pace

def add_km_markers(gdf_smooth, m):
    """Crculos blancos con nmero (halo) para cada kilmetro."""
    gdfm = gdf_smooth.to_crs(32719).copy()
    gdfm["segment_len_m"] = gdfm.length
    gdfm["dist_acum_m"] = gdfm["segment_len_m"].cumsum()

    kms = np.arange(1000, gdfm["dist_acum_m"].max(), 1000)  # 1 km, 2 km, ...
    marks = []
    for k in kms:
        idx = (gdfm["dist_acum_m"] - k).abs().argmin()
        seg = gdfm.iloc[idx].geometry
        pt_m = seg.interpolate(0.5, normalized=True)  # punto medio del segmento
        marks.append((int(k/1000), pt_m))

    gdf_marks = gpd.GeoDataFrame({"km":[km for km,_ in marks]}, geometry=[pt for _,pt in marks], crs=gdfm.crs).to_crs(4326)

    for _, row in gdf_marks.iterrows():
        lat, lon = row.geometry.y, row.geometry.x

        folium.CircleMarker(
            location=[lat, lon],
            radius=12, weight=2, color="#111",
            fill=True, fill_color="white", fill_opacity=0.95,
            popup=f"Km {int(row.km)}",
        ).add_to(m)

        # nmero con halo para mxima legibilidad
        folium.Marker(
            location=[lat, lon],
            icon=folium.DivIcon(
                html=f"""
                <div style="
                    font-size:14px;
                    font-weight:800;
                    color:#111;
                    text-align:center;
                    transform: translate(-50%, -55%);
                    text-shadow:
                        -1px -1px 0 #fff,
                         1px -1px 0 #fff,
                        -1px  1px 0 #fff,
                         1px  1px 0 #fff;
                ">{int(row.km)}</div>
                """
            ),
        ).add_to(m)
    return m

# ------------------ 1) LEER GPX ------------------
with open(GPX_FILE, "r") as f:
    gpx = gpxpy.parse(f)

seg = gpx.tracks[0].segments[0]
pts = seg.points
coords = [(p.longitude, p.latitude) for p in pts]

gdf_line  = gpd.GeoDataFrame({"name": ["Ruta 10K"]}, geometry=[LineString(coords)], crs="EPSG:4326")
gdf_start = gpd.GeoDataFrame({"label": ["Inicio"]}, geometry=[Point(coords[0])], crs="EPSG:4326")
gdf_end   = gpd.GeoDataFrame({"label": ["Fin"]},    geometry=[Point(coords[-1])], crs="EPSG:4326")

# ------------------ 2) SEGMENTOS + SUAVIZADO ------------------
gdf_seg    = build_segments_with_pace(pts)
gdf_smooth = smooth_by_target_meters(gdf_seg, target_m=SMOOTH_M)  # tiene 'avg_pace'

# ------------------ 3) HEATMAP POR CELDAS (bins humanos) ------------------
grid_pace = grid_heatmap_along_route_fixedbins(gdf_smooth, grid_m=GRID_M, buffer_m=BUFFER_M)
grid_pace_clean = grid_pace.dropna(subset=["pace_cell"]).copy()

m2 = grid_pace_clean.explore(
    tiles=TILE,
    column="pace_band",
    cmap="plasma",
    legend=True,
    style_kwds={"weight": 0.1, "fillOpacity": 0.55},
    name="Pace por zona (plasma)"
)
# ruta + inicio/fin
gdf_smooth.explore(m=m2, color="#111", style_kwds={"weight": 4, "opacity": 0.95}, name="Ruta")
gdf_start.explore(m=m2, marker_type="marker", color="green", name="Inicio")
gdf_end.explore(m=m2, marker_type="marker", color="red", name="Fin")

# encuadre y guardado
minx, miny, maxx, maxy = grid_pace_clean.total_bounds
m2.fit_bounds([[miny, minx], [maxy, maxx]])
m2.save(OUT_HTML_HEATMAP)

# ------------------ 4) MARCADORES DE KM + CONTROLES ------------------
# Creamos un grupo para los marcadores de km
layer_km = folium.FeatureGroup(name="Marcadores de km", show=True).add_to(m2)

m_final = add_km_markers(gdf_smooth, m2)
Fullscreen(position="topleft").add_to(m_final)
m_final.add_child(MeasureControl(position="topleft", primary_length_unit="meters"))
LayerControl(position="bottomleft", collapsed=False).add_to(m_final)

m_final.save(OUT_HTML_FINAL)
m_final
```
