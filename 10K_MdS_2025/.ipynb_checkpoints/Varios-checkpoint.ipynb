<b>Transformar de .gpx a geoJson</b>

```python
import gpxpy
import geojson

# Ruta al archivo GPX
gpx_file_path = 'activity_19162156358.gpx'

# Leer y parsear el archivo GPX
with open(gpx_file_path, 'r') as gpx_file:
    gpx = gpxpy.parse(gpx_file)

features = []

for track in gpx.tracks:
    for segment in track.segments:
        coords = []
        for point in segment.points:
            coords.append((point.longitude, point.latitude))  # Puedes agregar point.elevation si quieres

        feature = geojson.Feature(
            geometry=geojson.LineString(coords),
            properties={"name": track.name or "GPX Track"}
        )
        features.append(feature)

# Crear el objeto FeatureCollection
feature_collection = geojson.FeatureCollection(features)

# Guardar en archivo GeoJSON
geojson_path = gpx_file_path.replace('.gpx', '.geojson')
with open(geojson_path, 'w') as f:
    geojson.dump(feature_collection, f, indent=2)

print(f"Archivo GeoJSON creado: {geojson_path}")
```

```python
import gpxpy
import matplotlib.pyplot as plt

def extract_elevation_gpx(path):
    """Devuelve la lista de elevaciones de un archivo .gpx."""
    with open(path, 'r') as f:
        gpx = gpxpy.parse(f)

    elev = []
    for track in gpx.tracks:
        for segment in track.segments:
            for point in segment.points:
                if point.elevation is not None:
                    elev.append(point.elevation)
    return elev

# Rutas a tus sesiones
file1 = 'activity_19162156358.gpx'
file2 = 'activity_19936517952.gpx'

elev1 = extract_elevation_gpx(file1)
elev2 = extract_elevation_gpx(file2)

# Graficar
plt.figure(figsize=(12, 6))
plt.plot(elev1, label='MdS 2025', linewidth=2)
plt.plot(elev2, label='ASICS 2025', linewidth=2)
plt.title('Comparación de altimetría entre sesiones')
plt.xlabel('Puntos del recorrido')
plt.ylabel('Elevación (m)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# pip install gpxpy   # si no lo tienes aún
import gpxpy
import matplotlib.pyplot as plt

def perfil_gpx(path):
    """Devuelve (dist_km, elev_m) a partir de un archivo .gpx."""
    with open(path, 'r') as f:
        gpx = gpxpy.parse(f)

    dist_km, elev_m = [], []
    total_m = 0.0
    last = None

    for track in gpx.tracks:
        for seg in track.segments:
            for p in seg.points:
                # distancia acumulada
                if last is not None:
                    d = last.distance_3d(p)  # usa 3D si hay elevación, si no, cae a 2D
                    if d is None:
                        d = 0.0
                    total_m += d
                dist_km.append(total_m / 1000.0)

                # elevación
                elev_m.append(p.elevation if p.elevation is not None else elev_m[-1] if elev_m else 0.0)

                last = p
    return dist_km, elev_m

# Tus archivos
file1 = 'activity_19162156358.gpx'
file2 = 'activity_19936517952.gpx'

d1, e1 = perfil_gpx(file1)
d2, e2 = perfil_gpx(file2)

plt.figure(figsize=(12, 6))
plt.plot(d1, e1, label='MdS 2025', linewidth=2)
plt.plot(d2, e2, label='ASICS 2025', linewidth=2)
plt.title('Comparación de altimetría entre sesiones')
plt.xlabel('Distancia (km)')          # ← ahora en km
plt.ylabel('Elevación (m)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# pip install gpxpy
import gpxpy
import matplotlib.pyplot as plt
import numpy as np
from datetime import timedelta

def perfil_gpx(path, smooth=True, win=5):
    """
    Devuelve:
      dist_km  → distancia acumulada
      elev_m   → elevación
      speed_kh → velocidad instantánea (km/h)
    """
    with open(path, 'r') as f:
        gpx = gpxpy.parse(f)

    dist_km, elev_m, speed_kh = [], [], []
    total_m = 0.0
    last = None

    for trk in gpx.tracks:
        for seg in trk.segments:
            for p in seg.points:
                # distancia
                if last:
                    d = last.distance_3d(p) or 0.0
                    total_m += d
                    # tiempo
                    dt = (p.time - last.time).total_seconds() if p.time and last.time else 0.0
                    if dt > 0:
                        speed_kh.append( (d/1000.0) / (dt/3600.0) )  # km / h
                    else:
                        speed_kh.append(speed_kh[-1] if speed_kh else 0.0)
                else:
                    speed_kh.append(0.0)

                dist_km.append(total_m / 1000.0)
                elev_m.append(p.elevation if p.elevation is not None else elev_m[-1] if elev_m else 0.0)
                last = p

    if smooth and len(speed_kh) >= win:
        kernel = np.ones(win)/win
        speed_kh = np.convolve(speed_kh, kernel, mode='same')

    return dist_km, elev_m, speed_kh


file1 = 'activity_19162156358.gpx'
file2 = 'activity_19936517952.gpx'

d1, e1, s1 = perfil_gpx(file1)
d2, e2, s2 = perfil_gpx(file2)

fig, ax1 = plt.subplots(figsize=(12,6))

# Elevación
ax1.plot(d1, e1, label='Elevación MdS 2025', linewidth=2, color='tab:blue')
ax1.plot(d2, e2, label='Elevación ASICS 2025', linewidth=2, color='tab:orange')
ax1.set_xlabel('Distancia (km)')
ax1.set_ylabel('Elevación (m)')
ax1.grid(True)

# Velocidad
ax2 = ax1.twinx()
ax2.plot(d1, s1, label='Velocidad MdS 2025', linewidth=1.5, linestyle='--', color='tab:cyan')
ax2.plot(d2, s2, label='Velocidad ASICS 2025', linewidth=1.5, linestyle='--', color='tab:red')
ax2.set_ylabel('Velocidad (km/h)')

# Leyendas (una por eje)
lns = ax1.get_lines() + ax2.get_lines()
labels = [l.get_label() for l in lns]
ax1.legend(lns, labels, loc='upper center')

plt.title('Elevación y velocidad a lo largo del recorrido')
plt.tight_layout()
plt.show()
```

```python
# pip install gpxpy
import gpxpy
import matplotlib.pyplot as plt
import numpy as np

def perfil_gpx(path, smooth=True, win=5):
    """
    Devuelve:
      dist_km  → distancia acumulada
      elev_m   → elevación
      pace_min → ritmo instantáneo (min/km)
    """
    with open(path, 'r') as f:
        gpx = gpxpy.parse(f)

    dist_km, elev_m, pace_min = [], [], []
    total_m = 0.0
    last = None

    for trk in gpx.tracks:
        for seg in trk.segments:
            for p in seg.points:
                if last:
                    d = last.distance_3d(p) or 0.0            # metros
                    dt = (p.time - last.time).total_seconds() if p.time and last.time else 0.0
                    total_m += d

                    # ---- RITMO (min/km) ----
                    if d > 0 and dt > 0:
                        pace_min.append( (dt / 60.0) / (d / 1000.0) )
                    else:
                        pace_min.append(pace_min[-1] if pace_min else np.nan)
                else:
                    pace_min.append(np.nan)

                dist_km.append(total_m / 1000.0)
                elev_m.append(p.elevation if p.elevation is not None else elev_m[-1] if elev_m else np.nan)
                last = p

    # Suavizado opcional para el ritmo (media móvil)
    if smooth and len(pace_min) >= win:
        kernel = np.ones(win) / win
        pace_min = np.convolve(pace_min, kernel, mode='same')

    return dist_km, elev_m, pace_min


# Archivos
file1 = 'activity_19162156358.gpx'
file2 = 'activity_19936517952.gpx'

d1, e1, p1 = perfil_gpx(file1)
d2, e2, p2 = perfil_gpx(file2)

fig, ax1 = plt.subplots(figsize=(12,6))

# Elevación
ax1.plot(d1, e1, label='Elevación MdS 2025', linewidth=2, color='tab:blue')
ax1.plot(d2, e2, label='Elevación ASICS 2025', linewidth=2, color='tab:orange')
ax1.set_xlabel('Distancia (km)')
ax1.set_ylabel('Elevación (m)')
ax1.grid(True)

# Ritmo
ax2 = ax1.twinx()
ax2.plot(d1, p1, label='Ritmo MdS 2025', linewidth=1.5, linestyle='--', color='tab:cyan')
ax2.plot(d2, p2, label='Ritmo ASICS 2025', linewidth=1.5, linestyle='--', color='tab:red')
ax2.set_ylabel('Ritmo (min/km)')
ax2.invert_yaxis()          # ↓ = más rápido
ax2.set_ylim(ax2.get_ylim())  # fuerza actualización tras invertir

# Leyenda combinada
lns = ax1.get_lines() + ax2.get_lines()
ax1.legend(lns, [l.get_label() for l in lns], loc='upper center')

plt.title('Elevación y ritmo a lo largo del recorrido')
plt.tight_layout()
plt.show()
```

```python
import gpxpy
import numpy as np
import matplotlib.pyplot as plt

# ---------- helpers ----------

def load_track(path):
    """Devuelve distancia acumulada (km), elevación (m) y ritmo (min/km) por punto."""
    with open(path, 'r') as f:
        gpx = gpxpy.parse(f)

    dist_km, elev_m, pace_min = [], [], []
    total_m = 0.0
    last = None

    for trk in gpx.tracks:
        for seg in trk.segments:
            for p in seg.points:
                if last:                                    # distancia y tiempo entre puntos
                    d = last.distance_3d(p) or 0.0          # m
                    dt = (p.time - last.time).total_seconds() if p.time and last.time else 0.0
                    total_m += d
                    pace = (dt/60)/(d/1000) if d>0 and dt>0 else np.nan
                    pace_min.append(pace)
                else:
                    pace_min.append(np.nan)

                dist_km.append(total_m/1000)
                elev_m.append(p.elevation if p.elevation is not None else np.nan)
                last = p
    return np.array(dist_km), np.array(elev_m), np.array(pace_min)


def by_kilometer(dist_km, elev_m, pace_min):
    """Agrupa los tres vectores en promedios por cada km."""
    km_idx = np.floor(dist_km).astype(int)
    max_k = km_idx[-1]
    kms, pace_km, elev_km = [], [], []
    for k in range(max_k+1):
        sel = km_idx == k
        if sel.any():
            kms.append(k)
            elev_km.append(np.nanmean(elev_m[sel]))
            pace_km.append(np.nanmean(pace_min[sel]))
    return np.array(kms), np.array(pace_km), np.array(elev_km)


# ---------- carga y agregación ----------

file_gpx = 'activity_19162156358.gpx'        # cambia aquí tu GPX
dist, elev, pace = load_track(file_gpx)
kms, pace_km, elev_km = by_kilometer(dist, elev, pace)

# ---------- plot ----------

fig, ax1 = plt.subplots(figsize=(10,5))

# ritmo
ax1.plot(kms, pace_km, marker='o', color='tab:blue', label='Ritmo')
ax1.set_xlabel('Kilómetro')
ax1.set_ylabel('Ritmo (min/km)', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')
ax1.invert_yaxis()                   # ↓ = más rápido
ax1.grid(True, which='both', axis='both', linestyle=':')

# elevación
ax2 = ax1.twinx()
ax2.plot(kms, elev_km, marker='s', linestyle='--',
         color='gray', label='Elevación prom.')
ax2.set_ylabel('Elevación promedio (m)', color='gray')
ax2.tick_params(axis='y', labelcolor='gray')

plt.title('Ritmo vs Elevación por kilómetro')
fig.tight_layout()
plt.show()
```

```python
# pip install gpxpy
import gpxpy
import numpy as np
import matplotlib.pyplot as plt

# ---------- funciones reutilizables ----------

def load_track(path):
    """Devuelve dist_km, elev_m y pace_min POR PUNTO."""
    with open(path, 'r') as f:
        gpx = gpxpy.parse(f)

    dist_km, elev_m, pace_min = [], [], []
    total_m, last = 0.0, None

    for trk in gpx.tracks:
        for seg in trk.segments:
            for p in seg.points:
                if last:                                # distancia y tiempo entre puntos
                    d = last.distance_3d(p) or 0.0      # metros
                    dt = (p.time - last.time).total_seconds() if p.time and last.time else 0.0
                    total_m += d
                    pace = (dt/60) / (d/1000) if d>0 and dt>0 else np.nan
                    pace_min.append(pace)
                else:
                    pace_min.append(np.nan)

                dist_km.append(total_m/1000)
                elev_m.append(p.elevation if p.elevation is not None else np.nan)
                last = p
    return np.array(dist_km), np.array(elev_m), np.array(pace_min)


def by_kilometer(dist_km, elev_m, pace_min):
    """Promedia ritmo y elevación dentro de cada km completo."""
    km_idx = np.floor(dist_km).astype(int)          # 0-based: 0-0.999 km = km 0, etc.
    kms, pace_km, elev_km = [], [], []
    for k in range(km_idx[-1] + 1):
        sel = km_idx == k
        if sel.any():
            kms.append(k)
            pace_km.append(np.nanmean(pace_min[sel]))
            elev_km.append(np.nanmean(elev_m[sel]))
    return np.array(kms), np.array(pace_km), np.array(elev_km)


# ---------- carga de las dos sesiones ----------

file1 = 'activity_19162156358.gpx'
file2 = 'activity_19936517952.gpx'

d1, e1, p1 = load_track(file1)
d2, e2, p2 = load_track(file2)

kms1, pace1_km, elev1_km = by_kilometer(d1, e1, p1)
kms2, pace2_km, elev2_km = by_kilometer(d2, e2, p2)

# ---------- gráfico ----------

fig, ax1 = plt.subplots(figsize=(10,5))

# Ritmos
ax1.plot(kms1, pace1_km, marker='o', color='tab:blue', label='Ritmo MdS')
ax1.plot(kms2, pace2_km, marker='o', color='tab:orange', label='Ritmo ASICS')
ax1.set_xlabel('Kilómetro')
ax1.set_ylabel('Ritmo (min/km)')
ax1.invert_yaxis()                   # ↓ = más rápido
ax1.grid(True, which='both', axis='both', linestyle=':')

# Elevaciones (eje secundario)
ax2 = ax1.twinx()
ax2.plot(kms1, elev1_km, marker='s', linestyle='--', color='tab:blue', alpha=0.5,
         label='Elevación MdS')
ax2.plot(kms2, elev2_km, marker='s', linestyle='--', color='tab:orange', alpha=0.5,
         label='Elevación ASICS')
ax2.set_ylabel('Elevación promedio (m)')

# Leyenda combinada
lns = ax1.get_lines() + ax2.get_lines()
ax1.legend(lns, [l.get_label() for l in lns], loc='upper center', ncol=2)

plt.title('Ritmo y elevación por kilómetro (dos sesiones)')
plt.tight_layout()
plt.show()
```

```python
import gpxpy
import geojson
import os

# Ruta local del archivo GPX
gpx_file_path = 'Ingrid_Solis_2021-11-07_12-01-50.GPX'

# Leer y parsear el contenido del archivo GPX como texto
with open(gpx_file_path, 'r', encoding='utf-8') as gpx_file:
    gpx_content = gpx_file.read()
    gpx = gpxpy.parse(gpx_content)

# Crear las features desde los tracks
features = []
for track in gpx.tracks:
    for segment in track.segments:
        coords = [(point.longitude, point.latitude) for point in segment.points]
        feature = geojson.Feature(
            geometry=geojson.LineString(coords),
            properties={"name": track.name or "GPX Track"}
        )
        features.append(feature)

# Crear el FeatureCollection y guardar como .geojson
feature_collection = geojson.FeatureCollection(features)
geojson_path = os.path.splitext(gpx_file_path)[0] + '.geojson'

with open(geojson_path, 'w', encoding='utf-8') as f:
    geojson.dump(feature_collection, f, indent=2)

print(f"✅ Archivo GeoJSON creado: {geojson_path}")

```
