```python
# ============================================
# Ejemplo simple de uso de TensorFlow + Keras
# RegresiÃ³n: aprender y â 2x + 1
# ============================================

# 1. Importar librerÃ­as
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

print("VersiÃ³n de TensorFlow:", tf.__version__)

# 2. Generar datos sintÃ©ticos (x, y)
#    y = 2x + 1 + ruido
np.random.seed(42)

X = np.linspace(-1, 1, 200)              # 200 puntos entre -1 y 1
y = 2 * X + 1 + np.random.normal(0, 0.1, size=X.shape)  # lÃ­nea con ruido

# Keras espera matrices 2D: (n_muestras, n_features)
X = X.reshape(-1, 1)
y = y.reshape(-1, 1)

# 3. Definir el modelo en Keras (API de alto nivel de TensorFlow)
#    Modelo muy simple: una red con:
#      - Capa densa oculta con 8 neuronas (ReLU)
#      - Capa de salida con 1 neurona (predice y)
model = tf.keras.Sequential([
    tf.keras.layers.Dense(8, activation="relu", input_shape=(1,)),
    tf.keras.layers.Dense(1)
])

# Ver la arquitectura del modelo
model.summary()

# 4. Compilar el modelo
#    - Optimizador: Adam
#    - FunciÃ³n de pÃ©rdida: MSE (error cuadrÃ¡tico medio)
#    - MÃ©trica: MAE (error absoluto medio)
model.compile(
    optimizer="adam",
    loss="mse",
    metrics=["mae"]
)

# 5. Entrenar el modelo
history = model.fit(
    X, y,
    epochs=200,
    batch_size=16,
    verbose=0  # pon verbose=1 si quieres ver el progreso
)

# 6. Graficar la funciÃ³n real vs datos y vs predicciÃ³n del modelo
#    Creamos una malla de puntos para graficar una lÃ­nea suave
X_plot = np.linspace(-1.2, 1.2, 200).reshape(-1, 1)
y_true = 2 * X_plot + 1
y_pred = model.predict(X_plot)

plt.figure(figsize=(8, 5))
plt.scatter(X, y, label="Datos con ruido", alpha=0.6)
plt.plot(X_plot, y_true, label="FunciÃ³n real: y = 2x + 1")
plt.plot(X_plot, y_pred, label="Modelo TensorFlow", linestyle="--")
plt.xlabel("x")
plt.ylabel("y")
plt.title("RegresiÃ³n simple con TensorFlow")
plt.legend()
plt.grid(True)
plt.show()

# 7. Graficar la curva de pÃ©rdida durante el entrenamiento
plt.figure(figsize=(8, 4))
plt.plot(history.history["loss"])
plt.xlabel("Ãpoca")
plt.ylabel("PÃ©rdida (MSE)")
plt.title("EvoluciÃ³n de la pÃ©rdida durante el entrenamiento")
plt.grid(True)
plt.show()

# 8. Probar el modelo con algunos valores concretos
X_test = np.array([[-1.0], [0.0], [1.0], [2.0]])
y_test_true = 2 * X_test + 1
y_test_pred = model.predict(X_test)

print("Predicciones del modelo:")
for x_val, y_real, y_hat in zip(X_test, y_test_true, y_test_pred):
    print(f"x = {x_val[0]:4.1f} | y real = {y_real[0]:5.2f} | y predicho = {y_hat[0]:5.2f}")
```
